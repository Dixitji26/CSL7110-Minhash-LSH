```python
def read_file(filename):
    with open(filename, 'r') as f:
        return f.read().strip().lower()

docs = {
    "D1": read_file("D1.txt"),
    "D2": read_file("D2.txt"),
    "D3": read_file("D3.txt"),
    "D4": read_file("D4.txt"),
}

docs
```

```python
# Character k-grams
def char_kgrams(text, k):
    grams = set()
    for i in range(len(text) - k + 1):
        grams.add(text[i:i+k])
    return grams

# Word k-grams
def word_kgrams(text, k):
    words = text.split()
    grams = set()
    for i in range(len(words) - k + 1):
        grams.add(" ".join(words[i:i+k]))
    return grams
```

```python
char2 = {}
char3 = {}
word2 = {}

for name, text in docs.items():
    char2[name] = char_kgrams(text, 2)
    char3[name] = char_kgrams(text, 3)
    word2[name] = word_kgrams(text, 2)

print("k-grams generated successfully")
```

```python
def jaccard(set1, set2):
    intersection = len(set1 & set2)
    union = len(set1 | set2)
    return intersection / union if union != 0 else 0
```

```python
from itertools import combinations

pairs = list(combinations(docs.keys(), 2))

print("===== Character 2-grams =====")
for d1, d2 in pairs:
    print(f"{d1}-{d2}: {jaccard(char2[d1], char2[d2]):.4f}")

print("\n===== Character 3-grams =====")
for d1, d2 in pairs:
    print(f"{d1}-{d2}: {jaccard(char3[d1], char3[d2]):.4f}")

print("\n===== Word 2-grams =====")
for d1, d2 in pairs:
    print(f"{d1}-{d2}: {jaccard(word2[d1], word2[d2]):.4f}")
```

```python
import random

def generate_hash_functions(t, m=20011):
    hash_funcs = []
    for _ in range(t):
        a = random.randint(1, m-1)
        b = random.randint(0, m-1)
        hash_funcs.append((a, b))
    return hash_funcs
```

```python
def minhash_signature(grams, hash_funcs, m=20011):
    signature = []
    
    for a, b in hash_funcs:
        min_val = float('inf')
        
        for gram in grams:
            x = hash(gram) % m
            h = (a * x + b) % m
            if h < min_val:
                min_val = h
                
        signature.append(min_val)
    
    return signature
```

```python
def signature_similarity(sig1, sig2):
    matches = 0
    for i in range(len(sig1)):
        if sig1[i] == sig2[i]:
            matches += 1
    return matches / len(sig1)
```

```python
t_values = [20, 60, 150, 300, 600]

print("Exact Jaccard (3-grams) D1-D2:",
      jaccard(char3["D1"], char3["D2"]))

print("\nApproximate Jaccard using MinHash:")

for t in t_values:
    hash_funcs = generate_hash_functions(t)
    
    sig1 = minhash_signature(char3["D1"], hash_funcs)
    sig2 = minhash_signature(char3["D2"], hash_funcs)
    
    approx_sim = signature_similarity(sig1, sig2)
    
    print(f"t = {t}: {approx_sim:.4f}")
```

```python
t = 160
hash_funcs = generate_hash_functions(t)

sig1 = minhash_signature(char3["D1"], hash_funcs)
sig2 = minhash_signature(char3["D2"], hash_funcs)
sig3 = minhash_signature(char3["D3"], hash_funcs)
sig4 = minhash_signature(char3["D4"], hash_funcs)

signatures = {
    "D1": sig1,
    "D2": sig2,
    "D3": sig3,
    "D4": sig4
}
```

```python
def lsh_candidates(signatures, r, b):
    bands = [{} for _ in range(b)]
    
    for doc_name, sig in signatures.items():
        for band in range(b):
            start = band * r
            end = start + r
            band_slice = tuple(sig[start:end])
            
            if band_slice not in bands[band]:
                bands[band][band_slice] = []
            bands[band][band_slice].append(doc_name)
    
    candidates = set()
    
    for band in bands:
        for docs in band.values():
            if len(docs) > 1:
                for i in range(len(docs)):
                    for j in range(i+1, len(docs)):
                        candidates.add(tuple(sorted([docs[i], docs[j]])))
    
    return candidates
```

```python
from collections import defaultdict

user_movies = defaultdict(set)

with open("ml-100k/u.data", "r") as f:
    for line in f:
        user, movie, rating, _ = line.strip().split("\t")
        user_movies[int(user)].add(int(movie))

print("Total users:", len(user_movies))
```

```python
from itertools import combinations

def jaccard(set1, set2):
    return len(set1 & set2) / len(set1 | set2)

exact_similar_pairs = []

users = list(user_movies.keys())

for u1, u2 in combinations(users, 2):
    sim = jaccard(user_movies[u1], user_movies[u2])
    if sim >= 0.5:
        exact_similar_pairs.append((u1, u2))

print("Number of exact similar pairs (>=0.5):", len(exact_similar_pairs))
```

```python
import random

def generate_hash_functions(t, m=20011):
    hash_funcs = []
    for _ in range(t):
        a = random.randint(1, m-1)
        b = random.randint(0, m-1)
        hash_funcs.append((a, b))
    return hash_funcs
```

```python
def minhash_signature_movies(movies, hash_funcs, m=20011):
    signature = []
    
    for a, b in hash_funcs:
        min_val = float('inf')
        for movie in movies:
            h = (a * movie + b) % m
            if h < min_val:
                min_val = h
        signature.append(min_val)
    
    return signature
```

```python
def compute_signatures(t):
    hash_funcs = generate_hash_functions(t)
    signatures = {}
    
    for user in users:
        signatures[user] = minhash_signature_movies(user_movies[user], hash_funcs)
    
    return signatures
```

```python
def signature_similarity(sig1, sig2):
    return sum(1 for i in range(len(sig1)) if sig1[i] == sig2[i]) / len(sig1)
```

```python
def evaluate_minhash(t, runs=5):
    fp_total = 0
    fn_total = 0
    
    for _ in range(runs):
        signatures = compute_signatures(t)
        estimated_pairs = []
        
        for u1, u2 in combinations(users, 2):
            sim = signature_similarity(signatures[u1], signatures[u2])
            if sim >= 0.5:
                estimated_pairs.append((u1, u2))
        
        estimated_set = set(estimated_pairs)
        exact_set = set(exact_similar_pairs)
        
        false_positives = len(estimated_set - exact_set)
        false_negatives = len(exact_set - estimated_set)
        
        fp_total += false_positives
        fn_total += false_negatives
    
    print(f"\nt = {t}")
    print("Average False Positives:", fp_total / runs)
    print("Average False Negatives:", fn_total / runs)
```

```python
evaluate_minhash(50)
evaluate_minhash(100)
evaluate_minhash(200)
```

```python
def get_exact_pairs(threshold):
    exact_pairs = []
    
    for u1, u2 in combinations(users, 2):
        sim = jaccard(user_movies[u1], user_movies[u2])
        if sim >= threshold:
            exact_pairs.append((u1, u2))
    
    return set(exact_pairs)

exact_06 = get_exact_pairs(0.6)
exact_08 = get_exact_pairs(0.8)

print("Exact pairs ≥ 0.6:", len(exact_06))
print("Exact pairs ≥ 0.8:", len(exact_08))
```

```python
def lsh_candidates(signatures, r, b):
    bands = [{} for _ in range(b)]
    
    for user, sig in signatures.items():
        for band in range(b):
            start = band * r
            end = start + r
            band_slice = tuple(sig[start:end])
            
            if band_slice not in bands[band]:
                bands[band][band_slice] = []
            bands[band][band_slice].append(user)
    
    candidates = set()
    
    for band in bands:
        for users_list in band.values():
            if len(users_list) > 1:
                for i in range(len(users_list)):
                    for j in range(i+1, len(users_list)):
                        candidates.add(tuple(sorted((users_list[i], users_list[j]))))
    
    return candidates
```

```python
def evaluate_lsh(t, r, b, threshold, runs=5):
    fp_total = 0
    fn_total = 0
    
    for _ in range(runs):
        signatures = compute_signatures(t)
        candidates = lsh_candidates(signatures, r, b)
        
        if threshold == 0.6:
            exact_set = exact_06
        else:
            exact_set = exact_08
        
        false_positives = len(candidates - exact_set)
        false_negatives = len(exact_set - candidates)
        
        fp_total += false_positives
        fn_total += false_negatives
    
    print(f"\nConfiguration: t={t}, r={r}, b={b}, threshold={threshold}")
    print("Avg False Positives:", fp_total / runs)
    print("Avg False Negatives:", fn_total / runs)
```

```python
# For threshold 0.6
evaluate_lsh(50, 5, 10, 0.6)
evaluate_lsh(100, 5, 20, 0.6)
evaluate_lsh(200, 5, 40, 0.6)
evaluate_lsh(200, 10, 20, 0.6)

# For threshold 0.8
evaluate_lsh(50, 5, 10, 0.8)
evaluate_lsh(100, 5, 20, 0.8)
evaluate_lsh(200, 5, 40, 0.8)
evaluate_lsh(200, 10, 20, 0.8)
```
